Pandas is an **open source Python package** that is most widely used for data science/data analysis and machine learning tasks. # **Coorelations:**



```
A great aspect of the Pandas module is the corr() method.

The corr() method calculates the relationship between each column in your data set.
```



````python
data.corr()
````

**EXPLANATION OF RESULT (CORR):**


---


**NOTE:** The corr() method ignores "not numeric" columns.


```
The number varies from -1 to 1.

1 means that there is a 1 to 1 relationship (a perfect correlation), and for this data set, each time a value went up in the first column, the other one went up as well.

0.9 is also a good relationship, and if you increase one value, the other will probably increase as well.

-0.9 would be just as good relationship as 0.9, but if you increase one value, the other will probably go down.

0.2 means NOT a good relationship, meaning that if one value goes up does not mean that the other will.
```

**crosstab() function:**

# **PANDAS PROFILING:**



```

In short, what pandas profiling does is save us all the work of visualizing and understanding the distribution of each variable.
```

````python
# !pip install pandas-profiling==2.8.0
import pandas_profiling as pp
````

````python
report = data.profile_report(title = 'Data Analysis of Titanic Dataset')
````

````python
report
````

**What Pandas Profiling Covers:**

It Covers:

1. **Overview:** The Overview consists of overall statistics. This includes the number of variables (features or columns of the dataframe), Number of observations (rows of dataframe), Missing cells,  percentage of missing cells, Duplicate rows, percentage of duplicate rows, and Total size in memory.

 **Warnings:**The warnings tab consists of any type of warnings related to cardinality, correlation with other variables, missing values, zeroes, skewness of the variables, and many others.

 **Reproduction:**The reproduction tab simply displays information related to the report generation. It shows the start and ends the time of the analysis, the time taken to generate the report, the software version of pandas profiling, and a configuration download option.

2. Dataset info: Number of variables
Number of observations, Missing cells, Duplicate rows, Total size in memory and Average record size in memory

3. Variables: Missing values and its percentage, Distinct count, Unique
Correlations

4. Missing Values: ‘Matrix’ and ‘Count’

5. Sample: First 10 rows and Last 10 rows
